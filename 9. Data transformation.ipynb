```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/Users/SS/Downloads/Data Trasformation/adult_with_headers (1).csv")
df
```

```python
df.replace("?", np.nan, inplace=True)

df.dropna(inplace=True)

print("\nAfter handling missing values:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Standard Scaling
scaler_std = StandardScaler()
df_std = df.copy()
df_std[num_cols] = scaler_std.fit_transform(df[num_cols])

# Min-Max Scaling
scaler_mm = MinMaxScaler()
df_mm = df.copy()
df_mm[num_cols] = scaler_mm.fit_transform(df[num_cols])

print("\nStandard Scaled Data:\n", df_std.head())
print("\nMin-Max Scaled Data:\n", df_mm.head())
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# Identify categorical columns
cat_cols = df.select_dtypes(include='object').columns

# One-Hot Encoding (<5 categories)
for col in cat_cols:
    if df[col].nunique() < 5:
        df_encoded = pd.get_dummies(df_encoded, columns=[col], drop_first=True)

# Label Encoding (>5 categories)
le = LabelEncoder()
for col in cat_cols:
    if df[col].nunique() >= 5:
        df_encoded[col] = le.fit_transform(df[col])

print("\nEncoded Data:\n", df_encoded.head())
```
