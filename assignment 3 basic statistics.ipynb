```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

df = pd.read_csv("/Users/SS/Downloads/sales_data_with_discounts.csv")
df

```

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
print("Dataset Info:")
print(df.info())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64'])

print("\nDescriptive Statistics:")
print(num_cols.describe())

```

```python
# Mean
print("\nMean:\n", num_cols.mean())
```

```python
#median
print("\nMedian:\n", num_cols.median())
```

```python
#mode
print("\nMode:\n", num_cols.mode().iloc[0])
```

```python
#standard deviation
print("\nStandard Deviation:\n", num_cols.std())
```

```python
print("\nMissing Values:\n", df.isnull().sum())

```

```python
# Fill missing values with mean (for numerical columns)
df[num_cols.columns] = df[num_cols.columns].fillna(num_cols.mean())

```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
Data Visualization
●	Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
●	Histograms:
	Plot histograms for each numerical column.
	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.
●	Box Plots:
	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions.
●	Bar Chart Analysis for Categorical Column:
	Identify categorical columns in the dataset.
	Create bar charts to visualize the frequency or count of each category.
	Analyze the distribution of categories and provide insights
```

```python
# Histograms
for col in num_cols.columns:
    plt.figure()
    plt.hist(df[col], bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

```

```python
#boxplot
import seaborn as sns

for col in num_cols.columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#bar chat
cat_cols = df.select_dtypes(include=['object'])

for col in cat_cols.columns:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#heat plot
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
